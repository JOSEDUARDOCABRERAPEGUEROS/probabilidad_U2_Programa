```python
import math as mt # Para utilizar factorial
from itertools import permutations
from itertools import product
import numpy as np
```

```python
asistentes = ["a", "b", "c", "d", "e", "f", "g", "h", "i", "j"]
n = len(asistentes)
r = 4 # Representa nuestro tamao de subconjuntos

pr = mt.factorial(n) / mt.factorial(n-r)
print("Existen ", pr, " posibles permutaciones sin repeticin del total de ", n,
      " asistentes ", " en grupos de ", r, " en ", r)
print("Aqu las permutaciones")
permutacionesAsistentes = list(permutations(asistentes, r))

#Imprime las permutaciones
for permutacion in permutacionesAsistentes:
    print(permutacion)
```
